```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
b_0 = 0.05
angles = np.linspace(0,90,1000)
IAM_B = 1 - b_0 * ( (1 / np.cos(angles/180*np.pi)) - 1 )

fig, ax = plt.subplots(figsize=(5,4))

ax.plot(angles, IAM_B)

ax.set_xlabel('Incident Angle (deg)')
ax.set_ylabel('$IAM$')
ax.set_title('ASHRAE IAM Model')
ax.set_xlim(0,90)
ax.set_ylim(0,1.05)

fig.text(0.15, 0.15, r'$\mathbf{b_0=0.05}$', fontsize=9, color='#1f77b4')
#ax.text(0.5, 0.5, 'b_0=0.05', fontsize=9, fontweight='bold')


#plt.grid(True)
plt.show()
```

```python
import pandas as pd
import requests
api_url = 'https://re.jrc.ec.europa.eu/api/v5_2/tmy?'

params = {'lat': 60,
        'lon': 15, 
        'usehorizon': 1, 
        'raddatabase': 'PVGIS-SARAH2',
        'outputformat': 'json'}

response = requests.get(api_url, params=params)

if response.status_code == 200:
    print(response.json()['outputs'])
```
